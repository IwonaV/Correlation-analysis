```python
# Checking how much of the data is missing. 

missing_value = df_movies.isna().sum()/(len(df_movies))*100
missing_value
```

```python
# Changing the data type of some columns from float to integer.

cols = ['votes','budget','gross','runtime']
for col in cols:
   df_movies[col] = df_movies[col].apply(lambda x: int(x) if x == x else "")
```

```python
sorted_df_movies = df_movies.sort_values(by=['gross'],inplace= False, ascending=False)
```

```python
df_movies.drop_duplicates() 
# just in case but there was no duplicates
```

```python
# Checking if there is any relationship ( correlation) between budget and gross.
```

```python
plt.figure(figsize = (12, 8))
scatter = plt.scatter(x=df_movies['budget'],
            y=df_movies['gross'],
            alpha = 0.7,
            c = df_movies['budget'],
            cmap = 'mako_r')
plt.legend(*scatter.legend_elements(),
           loc = "upper left",
           title = "Budget")
plt.title("Relationship between Budget and Gross income", 
          fontsize = 14, 
          weight = "bold")
plt.xlabel("Gross income in USD", weight = "bold")
plt.ylabel("Budget in USD")
plt.show()
```
